```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
metrics_df = pd.DataFrame({
    "Subject": ["sub-1","sub-2","sub-3","sub-4","sub-5"],
    "Density": [0.34,0.29,0.37,0.31,0.28],
    "Clustering": [0.52,0.47,0.56,0.49,0.45],
    "Components": [1,1,1,2,2]
})

metrics_df["Topology Score"] = (
    metrics_df["Density"] + metrics_df["Clustering"]
)

metrics_df
```

```python
mean_score = metrics_df["Topology Score"].mean()
metrics_df["Deviation"] = metrics_df["Topology Score"] - mean_score

plt.figure(figsize=(7,4))
sns.barplot(data=metrics_df, x="Subject", y="Deviation")
plt.axhline(0, linestyle="--")
plt.title("Deviation from Mean Topology Score")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,2))

vals = metrics_df["Topology Score"]
plt.hlines(y=1, xmin=vals.min(), xmax=vals.max(), linewidth=3)
plt.scatter(vals, [1]*len(vals), s=100)

for _, row in metrics_df.iterrows():
    plt.text(row["Topology Score"], 1.02, row["Subject"], ha="center")

plt.yticks([])
plt.xlabel("Topology Score")
plt.title("Observed Topology Range Across Subjects")
plt.tight_layout()
plt.show()
```

```python
component_counts = metrics_df["Components"].value_counts().sort_index()

plt.figure(figsize=(6,4))
component_counts.plot(kind="bar")
plt.title("Connected Component Distribution")
plt.xlabel("Number of Components")
plt.ylabel("Subjects")
plt.tight_layout()
plt.show()
```

```python
roadmap = pd.DataFrame({
    "Stage": [
        "5 Subjects",
        "Larger Sample",
        "Vestibular ROI Focus",
        "Behavior Links",
        "Predictive Models"
    ],
    "Order": [1,2,3,4,5]
})

plt.figure(figsize=(8,4))
plt.plot(roadmap["Stage"], roadmap["Order"], marker="o")
plt.xticks(rotation=30, ha="right")
plt.title("Planned Scaling Path for Vestibular Repo 2")
plt.ylabel("Next Step Order")
plt.tight_layout()
plt.show()
```
