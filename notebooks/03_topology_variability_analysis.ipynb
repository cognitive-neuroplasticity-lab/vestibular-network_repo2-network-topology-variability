```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
metrics_df = pd.DataFrame({
    "Subject": ["sub-1","sub-2","sub-3","sub-4","sub-5"],
    "Density": [0.34,0.29,0.37,0.31,0.28],
    "Clustering": [0.52,0.47,0.56,0.49,0.45],
    "Components": [1,1,1,2,2]
})

metrics_df
```

```python
variability = pd.DataFrame({
    "Metric": ["Density","Clustering","Components"],
    "Mean": [
        metrics_df["Density"].mean(),
        metrics_df["Clustering"].mean(),
        metrics_df["Components"].mean()
    ],
    "SD": [
        metrics_df["Density"].std(),
        metrics_df["Clustering"].std(),
        metrics_df["Components"].std()
    ]
})

variability
```

```python
most_variable = variability.sort_values("SD", ascending=False)
most_variable
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=metrics_df, x="Density", y="Clustering", s=100)
plt.title("Density vs Clustering Across Subjects")
plt.tight_layout()
plt.show()
```

```python
metrics_df["Topology Score"] = (
    metrics_df["Density"] + metrics_df["Clustering"]
)

ranked = metrics_df.sort_values("Topology Score", ascending=False)
ranked[["Subject","Topology Score"]]
```
